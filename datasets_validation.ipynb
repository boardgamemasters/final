## DROP IRRELEVANT COLUMNS AND ROWS

```python
# Saves columns into a new dataframe. These columns will be transformed on later steps. 
totransform_df = gamesdb_3.loc[:, ['bgg_id', 'game_type', 'designer', 'artist','publisher', 'category', 
                                   'mechanic', 'family', 'implementation', 'integration']]
```

```python
# Saves compilations into a new dataframe. 
compilation_df = gamesdb_3.loc[gamesdb_3.compilation>0]
```

```python
# Copies gamesdb_3 in another variable. Drops rows without information for relevant dimensions
bgames = gamesdb_3.copy()
bgames.dropna(subset=['num_votes', 'year', 'category', 'mechanic', 'min_players'], inplace=True)
```

```python
# Drops columns without information or redundant one
bgames.drop(['bga_id', 'dbpedia_id', 'luding_id', 'spielen_id', 'wikidata_id', 'wikipedia_id', 
                'max_players_rec', 'max_players_best', 'min_age_rec', 'max_players', 'language_dependency', 
                'game_type', 'designer', 'artist','publisher', 'category', 'mechanic', 'family',
                'implementation', 'integration', 'compilation', 'compilation_of', 'bayes_rating'], axis=1, inplace=True)
```

```python
bgames.info()
```

## ADD DESCRIPTION, IMAGE PATH, URL

Missing values on new DB are URL, Description, ImagePath, NumExpansions, NumOwned, ,NumWant, NumWish, NumWeightVotes.
URL, Description & ImagePath will be added from our other DB since they are under any case influenced by time.
Important to identify what is an expansion or not in our new DB.

```python
# Merges gamesdb_3 with the other 2 DF in order to get columns for URL, Description and ImagePath

bgames = pd.merge(bgames, gamesdb_1[['ID', 'URL']], 
                     left_on='bgg_id', right_on='ID', how='left').drop('ID', axis=1)

bgames = pd.merge(bgames, gamesdb_2[['BGGId', 'Description', 'ImagePath']], 
                     left_on='bgg_id', right_on='BGGId', how='left').drop('BGGId', axis=1)
```

## BEAUTIFIER

```python
#Assumes NA as non cooperative games
bgames.cooperative=bgames.cooperative.fillna(0)
```

```python
bgames.info()
```

```python
# Reorganises columns in the DF
bgames= bgames.iloc[:,[0, 1, 2, 10, 11, 12, 13, 14, 3, 4, 5, 6, 7, 8, 9, 15, 16, 17]]
```

```python
# Standarises naming convention
bgames.rename({
    'URL':'url', 
    'Description':'description',
    'ImagePath':'image'}, axis=1, inplace=True)
```

```python
# Dictionary to convert the data types where possible
convert = {
    'bgg_id': 'int64',
    'name': 'object',
    'year': 'int64',
    'rank': 'float64',
    'num_votes': 'int64',
    'avg_rating': 'float64',
    'stddev_rating':'float64',
    'complexity': 'float64',
    'min_players': 'int64',
    'min_players_rec': 'int64',
    'min_players_best': 'int64',
    'min_age': 'float64',
    'min_time': 'float64',
    'max_time': 'float64',
    'cooperative': 'int64',
    'url': 'object',
    'description': 'object',
    'image': 'object', 
          }
```

```python
# Converts columns type based on the dictionary created beforehand
bgames = bgames.astype(convert)
```

## EXPORT TABLES

```python
bgames.to_csv("DB/DB/bgames_v1.csv", index = False)
compilation_df.to_csv("DB/DB/bgames.csv", index = False)
totransform_df.to_csv("DB/DB/totransform.csv", index = False)
```
