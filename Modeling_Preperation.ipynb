## Create ML Table

### category_df (min 100 Games to keep)

```python
keep_cat_id = (category_df.groupby('category_id').agg(amount=('bgg_id','count'))).copy()
keep_cat_id = keep_cat_id.loc[keep_cat_id['amount']>=100].reset_index()['category_id']
keep_cat_id
```

```python
cat_name   =   pd.read_csv("../data/third/bgg_Category.csv",low_memory=False)
```

```python
keep_cat=category_df.copy()
keep_cat=keep_cat.loc[keep_cat['category_id'].isin(keep_cat_id)]
keep_cat=keep_cat.merge(cat_name.rename({'bgg_id': 'category_id'}, axis=1), on='category_id', how='left')
keep_cat['val']=1
keep_cat
```

```python
merge_cat= (pd.pivot_table(
    keep_cat, values='val', index='bgg_id', columns='name', fill_value=0
).reset_index())
merge_cat
```

### family_df
First: Keep Families with at least sqrt of game_nr and than use IQR*1.5 to filter max amount

```python
min_filter=np.sqrt(len(game_frame['bgg_id'].unique()))
```

```python
keep_fam_id = (family_df.groupby('family_id').agg(amount=('bgg_id','count'))).copy()
keep_fam_id = keep_fam_id.loc[(keep_fam_id['amount']>=min_filter)]
min_keep = (keep_fam_id.describe()).iloc[4,0]-((keep_fam_id.describe()).iloc[6,0] - (keep_fam_id.describe()).iloc[4,0])
max_keep = (keep_fam_id.describe()).iloc[6,0] + ((keep_fam_id.describe()).iloc[6,0] - (keep_fam_id.describe()).iloc[4,0])
keep_fam_id.describe(),min_keep, max_keep
keep_fam_id = keep_fam_id.loc[(keep_fam_id['amount']>=min_keep) & (keep_fam_id['amount']<=max_keep)].reset_index()['family_id']
keep_fam_id
```

```python
fam_name   =   pd.read_csv("../data/third/bgg_GameFamily.csv",low_memory=False)
```

```python
keep_fam=family_df.copy()
keep_fam=keep_fam.loc[keep_fam['family_id'].isin(keep_fam_id)]
keep_fam=keep_fam.merge(fam_name.rename({'bgg_id': 'family_id'}, axis=1), on='family_id', how='left')
keep_fam['val']=1
keep_fam
```

```python
merge_fam= (pd.pivot_table(
    keep_fam, values='val', index='bgg_id', columns='name', fill_value=0
).reset_index())
merge_fam
```

### game_type_df
Keep all  ames_types with amount >1

```python
keep_gt_id = (game_type_df.groupby('game_type_id').agg(amount=('bgg_id','count'))).copy()
keep_gt_id = keep_gt_id.loc[(keep_gt_id['amount']>1)].reset_index()['game_type_id']
```

```python
gt_name   =   pd.read_csv("../data/third/bgg_GameType.csv",low_memory=False)
```

```python
keep_gt=game_type_df.copy()
keep_gt=keep_gt.loc[keep_gt['game_type_id'].isin(keep_gt_id)]
keep_gt=keep_gt.merge(gt_name.rename({'bgg_id': 'game_type_id'}, axis=1), on='game_type_id', how='left')
keep_gt['val']=1
```

```python
merge_gt= (pd.pivot_table(
    keep_gt, values='val', index='bgg_id', columns='name', fill_value=0
).reset_index())
merge_gt
```

### mechanic_df
treshold 96.5

```python
keep_mech_id = (mechanic_df.groupby('mechanic_id').agg(amount=('bgg_id','count'))).copy()
keep_mech_id = keep_mech_id.loc[(keep_mech_id['amount']>96.5)].reset_index()['mechanic_id']
```

```python
mech_name   =   pd.read_csv("../data/third/bgg_Mechanic.csv",low_memory=False)
```

```python
keep_mech=mechanic_df.copy()
keep_mech=keep_mech.loc[keep_mech['mechanic_id'].isin(keep_mech_id)]
keep_mech=keep_mech.merge(mech_name.rename({'bgg_id': 'mechanic_id'}, axis=1), on='mechanic_id', how='left')
keep_mech['val']=1
```

```python
merge_mech= (pd.pivot_table(
    keep_mech, values='val', index='bgg_id', columns='name', fill_value=0
).reset_index())
merge_mech
```

## Marry all resulting Tables

```python
game_learn_df = (
    game_frame.merge(
        merge_cat, on='bgg_id', how='left'
    ).merge(
        merge_fam, on='bgg_id', how='left'
    ).merge(
        merge_gt, on='bgg_id', how='left'
    ).merge(
        merge_mech, on='bgg_id', how='left'
    )
)
game_learn_df.describe()
```
